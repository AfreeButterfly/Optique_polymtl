```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

def classic1Pol(theta):
    E_0=np.sqrt(2*377)
    new_intensity=1/(2*377)*np.abs(E_0*np.cos(theta))**2
    return new_intensity

def classic2Pol(theta):
    
    nabla_2=2*377
    E_0=(nabla_2)**(3/4)
    E_1=np.linalg.norm(E_0*np.cos(theta))**2*1/nabla_2
    E_2=np.linalg.norm(E_1*np.cos(2*theta))**2*1/nabla_2

    return E_2

thetaPoints = np.linspace(0, np.pi/2, 100)
power1Coeff = []
power2Coeff = []


for theta in thetaPoints : 
    power1Coeff.append(classic1Pol(theta))
    power2Coeff.append(classic2Pol(theta))




plt.plot(thetaPoints, power1Coeff, label='2 polariseurs')
plt.plot(thetaPoints, power2Coeff, label='3 polariseurs')
plt.xlabel(f'$\Theta (rad)$')
plt.ylabel(f'T($\Theta$)')
plt.legend()


```
