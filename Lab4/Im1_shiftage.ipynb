```python
import numpy as np
import matplotlib.pyplot as plt
```

Show original image

```python
im1 = plt.imread('im1.png')

plt.figure(figsize=[10, 5])
# Show color image
plt.imshow(im1, cmap='gray')
plt.xlabel('x (j)')
plt.ylabel('y (i)')
plt.title('im1')
plt.savefig("IM1_pictures/im1_graph.png")
```

Show 2D fft of the image

```python
im1_freq = np.fft.fft2(im1) # compute fft2
im1_freq = np.fft.fftshift(im1_freq) # shift fft2

# plot results
plt.imshow(np.log(np.abs(im1_freq)),cmap="gray")
plt.title('im1 2d fft')
plt.savefig("IM1_pictures/im1_2dFFT_graph.png")
```

Show filtered FFT of the image

```python
w = np.zeros(im1.shape)   #+np.ones(im1.shape)/10e10
w[50,:]=1

im1_freq_filtered=im1_freq*w
plt.imshow(np.log(np.abs(im1_freq))*w,cmap="gray")
plt.title('im1 2d fft filtered')
plt.savefig("IM1_pictures/im1_2dfft_filtered_graph.png")
```

Show filtered image

```python
im1_filtered=np.fft.ifft2(np.fft.fftshift(im1_freq_filtered)).real

# plt.figure(figsize=[10, 5])
# # Show color image
# plt.subplot(1,2,1)
# plt.imshow(im1, cmap='gray')
# plt.xlabel('x (j)')
# plt.ylabel('y (i)')

# # Show greyscale image
# plt.subplot(1,2,2)
plt.imshow(im1_filtered, cmap="gray")
plt.xlabel('x (j)')
plt.ylabel('y (i)')
plt.title('im1 filtered')
plt.savefig("IM1_pictures/im1_filtered_graph.png")
```
