```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
im1 = plt.imread('im1.png')

plt.figure(figsize=[10, 5])
# Show color image
plt.imshow(im1, cmap='gray')
plt.xlabel('x (j)')
plt.ylabel('y (i)')
```

```python
im1_freq = np.fft.fft2(im1) # compute fft2
im1_freq = np.fft.fftshift(im1_freq) # shift fft2

# plot results
plt.imshow(np.log(np.abs(im1_freq)))
plt.title('im1 2d fft')
```

```python
im1_freq = np.fft.fft(im1) # compute fft2
im1_freq = np.fft.fftshift(im1_freq) # shift fft2

# plot results
plt.imshow(np.log(np.abs(im1_freq)))
plt.title('im1 2d fft')
```
